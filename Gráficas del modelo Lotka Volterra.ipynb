```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Cacomixtle y oecanthus niveus

```python
def df_dt(x, t, r, a, b, c, k):
    dx = (r * x[0]) * (1 - x[0]/k) - a * x[0] * x[1] 
    dy =  - c * x[1] + b * x[0] * x[1]
    return np.array([dx, dy])
# Parámetros
r = 3.0 #tasa de crecimiento presa
a = 1.4 #muerte de presa
b = 0.4 #tasa de crecimiento cacomixtle
c = 0.7 #muerte cacomixtle
k = 30.5 #capacidad de carga
# Condiciones iniciales
x0 = 157  # Presas
y0 = 54   # Depredadores
conds_iniciales = np.array([x0, y0])
# Condiciones para integración
tf = 500
N = 1000
t = np.linspace(0, tf, N)
solucion = odeint(df_dt, conds_iniciales, t, args=(r, a, b, c, k))
```

```python
plt.plot(t, solucion[:, 0], label='Oecanthus niveus')
plt.plot(t, solucion[:, 1], label='Bassaricus astutus')
plt.xlim(0, 150)
plt.ylim(0, 10)
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.grid()
plt.legend()
plt.show()
```

```python

plt.plot(solucion[:, 0], solucion[:, 1])
plt.xlim(0, 23)
plt.ylim(0, 10)
plt.xlabel('Oecantus californicus')
plt.ylabel('Bassarius astutus')
plt.legend()
plt.show()
```

# Urocyon cinereoargenteus y Oecanthus californicus

```python
def df_dt(x, t, r, a, b, c, k):
    dx = (r * x[0]) * (1 - x[0]/k) - a * x[0] * x[1] 
    dy =  - c * x[1] + b * x[0] * x[1]
    return np.array([dx, dy])
# Parámetros
r = 3.0 #tasa de crecimiento presa
a = 1.4 #muerte de presa
b = 0.3 #tasa de crecimiento de la zorra
c = 0.8 #muerte zorra
k = 66
# Condiciones iniciales
x0 = 78  # Presas
y0 = 1   # Depredadores
conds_iniciales = np.array([x0, y0])
# Condiciones para integración
tf = 500
N = 1000
t = np.linspace(0, tf, N)
dos_solucion = odeint(df_dt, conds_iniciales, t, args=(r, a, b, c, k))
```

```python
plt.plot(t, dos_solucion[:, 0], label='Oecanthus californicus')
plt.plot(t, dos_solucion[:, 1], label='Urocyon cinereoargenteus')
plt.xlim(0, 120)
plt.ylim(0, 20)
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.grid()
plt.legend()
plt.show()
```
